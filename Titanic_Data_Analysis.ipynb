##Q12.Show age frequency with survived and not survied passengers.(Histogram)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and already loaded with Titanic data
S = df[df['Survived'] == 1]['Age']

# Create the histogram
counts, bins, _ = plt.hist(S, color='olive', edgecolor="black", bins=15)

# Calculate the bin centers
bin_centre = (bins[:-1] + bins[1:]) / 2

# Plot the data points (using black circles) over the histogram
plt.plot(bin_centre, counts, marker='o', color='black')

# Customize the x-axis ticks
plt.xticks(np.arange(0, 84, 4))

# Add titles and labels
plt.title("Survived Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

##Q14. How many passengers are travelled in different class show in a bar.

```python
temp=df.Pclass.value_counts()
temp
```

```python
data = dict(temp)
x=data.keys()
count=data.values()
count
```

```python
import matplotlib.pyplot as plt
val=[]
for i in x:
    if i == 1:
        val.append("1st Class")
    elif i==2:
        val.append("2nd Class")
    elif i ==3:
        val.append("3rd Class")

plt.bar(val,count,color=["lightblue","pink","lightgreen"])
plt.title("Passengers Travelled in different Classes")
plt.xlabel("Classes")
plt.ylabel("No. of persons")
plt.show
```

##Q15 How many passengers are survived with class wise and show in Bar Graph.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'Pclass' and 'Survived' to count the number of survivors in each class
survival_counts = df.groupby(['Pclass', 'Survived']).size().unstack(fill_value=0)

# Rename the Survived values from 0 and 1 to 'Not Survived' and 'Survived'
survival_counts.columns = ['Not Survived', 'Survived']

# Plotting the bar graph
survival_counts.plot(kind='bar', stacked=False, color=['salmon', 'lightgreen'], edgecolor='black')

# Adding title and labels
plt.title('Survival Count by Class')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)  # To keep the x-axis labels horizontal
plt.show()
```
